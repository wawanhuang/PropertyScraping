# III. Scraping/Get Data

```python

# Inisialisasi list untuk menyimpan data
price =[]
lokasi = []
tipe = []
group = []
luas_bangunan = []
luas_tanah = []
sertifikat = []
furnish = []
garasi = []

# Inisialisasi WebDriver
# driver = webdriver.Chrome()

# Loop untuk memproses halaman
for pagee in range(1, 34):
    url = f"https://www.brighton.co.id/dijual/rumah/samarinda?Keyword=samarinda&OrderBy=5&page={pagee}"
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    wrapper_divs = soup.find_all('div', class_='col-12 col-md-6 col-lg-4 py-2 px-0 py-sm-2 px-md-2')

    # Proses setiap listing properti di halaman
    for wrapper in wrapper_divs:
        try:
            link = wrapper.find('a', class_="record-page-visitor")['href']
            driver.get(link)
            html_ind = driver.page_source
            soup_ind = BeautifulSoup(html_ind, "html.parser")

            # Temukan semua elemen div dengan class yang benar
            harga = soup_ind.find_all('div', class_='wrapper-price detail-property pb-2')[0].get_text()
            price.append(harga)
            # try:
            #     hargas = harga.find('div', {'class':'wrapper-price detail-property pb-2'})
            #     harga.append(hargas.get_text())
            # except:
            #     harga.append(None)
            info_divs = soup_ind.find_all('div', class_='wrapper-info-detail')  # Ganti dengan class yang benar
            # for hargas in harga:
            #     p_harga = hargas.find_all('p')
            #     for j in range(0, len(p_harga), 2):
            #         if j + 1 < len(p_harga):
            #             key = p_harga[i].text.strip()
            #             value = p_harga[i+1].text.strip()
            #             if key == 'Price':
            #                 lokasi.append(value)
            for info_div in info_divs:
                p_tags = info_div.find_all('p')
                # time.sleep(5)  # Tunggu 5 detik untuk memastikan halaman properti dimuat
                for i in range(0, len(p_tags), 2):
                    if i + 1 < len(p_tags):
                        key = p_tags[i].text.strip()
                        value = p_tags[i+1].text.strip()

                        # Append values to appropriate lists based on the key
                        if key == 'Lokasi':
                            lokasi.append(value)
                        elif key == 'Tipe':
                            tipe.append(value)
                        elif key == 'Group':
                            group.append(value)
                        elif key == 'Luas Bangunan':
                            luas_bangunan.append(value)
                        elif key == 'Luas Tanah':
                            luas_tanah.append(value)
                        elif key == 'Sertifikat':
                            sertifikat.append(value)
                        elif key == 'Furnish':
                            furnish.append(value)
                        elif key == 'Garasi':
                            garasi.append(value)

        except Exception as e:
            print(f"Error processing property at {link}: {e}")
            continue

# Tutup WebDriver
# driver.quit()

# Buat DataFrame
pages = pd.DataFrame()
pages['Price'] = price
pages['Lokasi'] = lokasi
pages['Tipe'] = tipe
pages['Group'] = group
pages['Luas Bangunan'] = luas_bangunan
pages['Luas Tanah'] = luas_tanah
pages['Sertifikat'] = sertifikat
pages['Furnish'] = furnish

pages
```

```python
samarinda = pages.to_csv('samarinda.csv', index=False)
```
